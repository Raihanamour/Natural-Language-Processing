### Exercise 1

#### 1. Extract all NOUN tokens from this story.
#### 2. Extract all numbers (NUM POS type) in a python list
#### 3. Print a count of all POS tags in this story


```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open('news_story.txt','r') as f:
    news_text = f.read()
    
```

```python
news_text[:500]
```

### Extract all NOUN tokens from this story

```python
doc = nlp(news_text)
```

```python
for token in doc:
    print(token,'|',token.pos_)
```

```python
noun = []
```

```python
for token in doc:
    if token.pos_ == 'NOUN':
        noun.append(token)
```

```python
noun
```

### Extract NOUN and NUM tokens

```python
doc = nlp(news_text)
```

```python
number_tokens = []
noun_tokens = []
```

```python
for token in doc:
    if token.pos_ == 'NOUN':
        noun_tokens.append(token)
```

```python
noun_tokens
```

```python
for token in doc:
    if token.pos_ == 'NUM':
        number_tokens.append(token)
```

```python
number_tokens
```

### Print a count of all POS tags in this story


```python
count = doc.count_by(spacy.attrs.POS)
```

```python
count
```

```python
for k,v in count.items():
    print(doc.vocab[k].text,v)
```
